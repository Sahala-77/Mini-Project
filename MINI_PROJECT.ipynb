```python
# New text data for prediction
new_texts = [
    "Excellent training sessions, well-structured and easy to understand.",
    "The training is decent, but more real-time projects would help."
]
```

```python
# Vectorize the new text data
new_X = vectorizer.transform(new_texts)
```

```python
# Predict the labels for new text data
predictions = classifier.predict(new_X)
```
